Seperating categorical and continous columns from the dataset

```python
cat = []
con = []
for i in A.columns:
    if (A[i].dtypes=='object'):
        cat.append(i)
    else:
        con.append(i)
```

```python
cat # categorical columns
```

```python
con #continuous columns
```

```python
#Treating missing values of continuous columns
X = A[['Rear.seat.room']].mean()
A["Rear.seat.room"]= A["Rear.seat.room"].fillna(X[0])
```

```python
X = A[["Luggage.room"]].mean()
A["Luggage.room"] =  A["Luggage.room"].fillna(X[0])
```

```python
# Treating missing values of categorical columns
X = A.AirBags.mode()
A.AirBags = A.AirBags.fillna(X[0])
```

```python
A.isna().sum()
```

```python
A = A.drop(labels=["Unnamed: 0"],axis=1)
```

```python
A = A.drop(labels=["Make","Model","Manufacturer"],axis=1)
```

Bivariate Analysis

```python
for i in A.columns:
    if (A[i].dtypes=="object"):
        sb.boxplot(A[i],A.Width) # boxplot between categorical and continuous
        plt.xlabel(i)
        plt.ylabel('Width')
        plt.show()
    else:
        plt.scatter(A[i],A.Width) # scatterplot between continuous and continuous
        plt.xlabel(i)
        plt.ylabel('Width')
        plt.show()
```
